## Maps / Data Manipulation / Data Transformations

Maps are a function that takes one set of values and "maps" them to another set of values. 

```python
#Basic example
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
#Applied to dataframe (notice points has changed from the original DF)
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
#Notice that the original DF has not changed
reviews.head(1)
```

```python
#Creating a new column that combines two different columns
reviews.country + " - " + reviews.region_1
```

## Grouping and Sorting

References

https://www.kaggle.com/residentmario/grouping-and-sorting-reference

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

https://www.kaggle.com/residentmario/grouping-and-sorting-reference

```python
#Frequency histogram
reviews.groupby('points').points.count()
```

```python
#Frequency histogram, cheapest wine per point value category
reviews.groupby('points').price.min()
```

```python
#selects the name of the first wine reviewed from each winery in the dataset
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
#you can take this further and group by more than one column (picks best wine by country and providence)
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.argmax()])
```

```python
#generating some summary statistics using agg
reviews.groupby(['country']).price.agg([len, min, max])
```

## Working with Multiple Indexes

When slicing up data for analysis using multiple indexes can help you really drill down to info

References:

https://pandas.pydata.org/pandas-docs/stable/advanced.html

```python
#using multiple indexes
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

## Sorting

```python
#See resetting index https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.reset_index.html
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
#Ascending=False reverses the order of the sort to highest to lowest
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
#Another way to sort (this one by index)
countries_reviewed.sort_index()
```

```python
#sort by multiple columns

countries_reviewed.sort_values(by=['country', 'len'])
```
