## Plotnine

Plotnine is a really polished visualization library. Use this when you need it to look really good. It's also really good at letting you refine and add things as you go. In the end it's a lot quicker than other options. Faceting is really much easier than other libraries. 

The general progression of plotnine is:

1. Define data
2. Setup asthetics
3. Add layers
4. Repeat 2 and 3 until it's how you want it

```python
#Define the data
import pandas as pd
reviews = pd.read_csv("data/winemag-data-130k-v2.csv", index_col=0)
reviews.head(3)
```

```python
from plotnine import *

top_wines = reviews[reviews['variety'].isin(reviews['variety'].value_counts().head(5).index)]
```

```python
df = top_wines.head(1000).dropna()

#setup a simple scatterplot
#notice the steps being used 
#1. data (ggplot(df))
#2. asthetics (aes)
#3. layers (geom_point())
(ggplot(df)
 + aes('points', 'price')
 + geom_point())
```

```python
#now add another layer: stat_smooth() to add the trend line

df = top_wines.head(1000).dropna()

(
    ggplot(df)
        + aes('points', 'price')
        + geom_point()
        + stat_smooth()
)
```

```python
#add some colors with aes

df = top_wines.head(1000).dropna()

(
    ggplot(df)
        + geom_point()
        + aes(color='points')
        + aes('points', 'price')
        + stat_smooth()
)
```

```python
#add some faceting with facet_wrap

df = top_wines.head(1000).dropna()

(ggplot(df)
     + aes('points', 'price')
     + aes(color='points')
     + geom_point()
     + stat_smooth()
     + facet_wrap('~variety')
)
```

```python
#barchart
(ggplot(top_wines)
     + aes('points')
     + geom_bar()
)
```

```python
#hexplot
(ggplot(top_wines)
     + aes('points', 'variety')
     + geom_bin2d(bins=20)
)
```

```python
(ggplot(top_wines)
         + aes('points', 'variety')
         + geom_bin2d(bins=20)
         + coord_fixed(ratio=1)
         + ggtitle("Top Five Most Common Wine Variety Points Awarded")
)
```
